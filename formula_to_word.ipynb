```python
def Errormsg(error):
    if error == "Import":
        print("-"*45)
        print("ImportError: pyperclip module is not installed.")
        print("-"*45)

    if error == "Type":
        print("-"*45)
        format = [
            "1. matrix H W num1 num2 ... numH*W",
            "2. emptymatrix H W",
            "3. ematrix N",
            "4. eqarray eq1,eq2,...,eqN"
        ]
        print("TypeError: Please check the input format.")
        print("format: " + "\n        ".join(format))
        print("Tip: You can change the parentheses of matrix by use format: func + prth. (ex. matrix||)")
        print(" "*5 + "parentheses: func[] or func|| or func|||| .   func() is default.")
    
    if error == "natural num":
        print("-"*45)
        print("InputError: The height and width of the matrix must be natural numbers.")

    if error == "Func":
        print("-"*45)
        print("FunctionError: The function you entered is not supported.")
        print(" "*15 + "Please check the input format.")
    
    if error == "Unexpected":
        print("-"*45)
        print("UnexpectedError: An unexpected error occurred.")
        print(" "*17 + "Please check for other errors first.")
        print(" "*17 + "If this error persists only, please report it to the developer with your input.")
        
def matrix_process(matrix,prth,mode):
    result, matrix_word, list_mtx = matrix_to_word(matrix,prth,mode)
    if result == True:
        view_matrix(list_mtx,prth,mode)
        print("-"*45)
        print(matrix_word)
        pc.copy(matrix_word)
        print("--- Matrix copied to clipboard. ---")

def matrix_to_word(matrix,prth,mode):
    if mode == "E":
        H,W = int(matrix),int(matrix)
        elements = []
    else:
        H,W,*elements = map(float,matrix.split(" "))
        H,W = int(H),int(W)

    if (len(elements) != H * W and mode == "normal") or (mode != "normal" and len(elements) !=0):
        Errormsg("Type")
        return False,""
    elif H <= 0 or W <= 0:
        Errormsg("natural num")
        return False,""
    else:
        list_mtx = list_matrix(H,W,elements,mode)
        inner = "@".join(["&".join(list(map(str,i))) for i in list_mtx])
        if mode == "empty":
            inner = inner.replace("0","")
        matrix_word = f"{prth[:len(prth)//2]}■({inner}){prth[len(prth)//2:]}"
        return True, matrix_word, list_mtx

def list_matrix(H,W,elements,mode):
    if mode == "normal":
        matrix = [list(map(float,elements[i*W:(i+1)*W])) for i in range(H)]
        matrix = [list(map(lambda x: int(x) if x.is_integer() else x, row)) for row in matrix]
    if mode == "empty":
        matrix = [[0]*W for _ in range(H)]
    if mode == "E":
        matrix = [[1 if i==j else 0 for j in range(W)] for i in range(H)]
    return matrix

def view_matrix(list_mtx,prth,mode):
    print("-"*45)
    if mode == "empty":
        print(f"Recognized {len(list_mtx)}*{len(list_mtx[0])} empty matrix.")
    else:
        print("View:")
        col_widths = [max(map(lambda x:len(str(x)), col)) for col in zip(*list_mtx)]
        for row in list_mtx:
            formatted_row = " ".join(f"{val:>{col_widths[i]}}" for i, val in enumerate(row))
            print(formatted_row)
    print(f"with: {prth}")
        
def eqarray_process(eqs):
    eqs = eqs.split(",")
    eqs = list(map(lambda x:x.replace(" ","").split("="), eqs))
    eqs_word = eqs_to_word(eqs)
    print("-"*45)
    print(eqs_word)
    pc.copy(eqs_word)
    print("--- Equations copied to clipboard. ---")

def eqs_to_word(eqs):
    inner = "@".join(["&=".join(i) for i in eqs])
    eqs_word = "{"+f"█({inner})┤"
    return eqs_word
```

```python
Error = False
complete = False
Unexpected = False

try:
    import pyperclip as pc
except:
    Errormsg("Import")
    Error = True
    complete = True

function = input("Enter a function: ")
print(f"Input: {function}")
func_list = ["matrix", "emptymatrix", "ematrix", "eqarray"]
#-----------------------------------------------------------------------------------------------------------
#matrix H W elements           # name + parentheses is allowed. Not need parentheses if you want to use ().
#emptymatrix H W               # (ex. matrix|| H W elements  --> |matrix|)
#Ematrix N                     # parentheses: [] OR || OR |||| 
#-----------------------------------------------------------------------------------------------------------
#eqarray eq1,eq2, ...
#-----------------------------------------------------------------------------------------------------------
try:
    func, inst = function.split(" ",1)
    func = func.replace("||||","‖‖")
    if not (func in func_list or func[:-2] in func_list):
        Errormsg("Func")
        Error = True
        complete = True
except:
    if " " not in function:
        Errormsg("Type")
        Error = True
        complete = True
    Unexpected = True
#-----------------------------------------------------------------------------------------------------------
try:
    if complete == False:
        if "matrix" in func:
            if func[-2:]!="ix":
                prth = func[-2:]
                func = func[:-2]
            else:
                prth = "()"

        if Error == False:
            matrix_mode = {"matrix":"normal", "emptymatrix":"empty", "ematrix":"E"}
            matrix_process(inst,prth,matrix_mode[func])
            complete = True
except:
    Unexpected = True
#-----------------------------------------------------------------------------------------------------------
try:
    if complete == False:
        if func == "eqarray":
            eqarray_process(inst)
            complete = True
except:
    Unexpected = True
#-----------------------------------------------------------------------------------------------------------
if complete == False and Unexpected == True:
    Errormsg("Unexpected")
```
